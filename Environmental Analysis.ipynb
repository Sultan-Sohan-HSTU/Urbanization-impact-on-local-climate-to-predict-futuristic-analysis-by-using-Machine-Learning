```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Historical data
years = np.array([2000, 2005, 2010, 2015, 2020, 2025]).reshape(-1, 1)
population = np.array([10.0, 12.0, 14.5, 17.0, 19.5, 21.0])  # in millions
vegetation = np.array([60, 55, 50, 45, 40, 35])  # in sq km
aqi = np.array([120, 140, 160, 180, 210, 230])  # Air Quality Index
temperature = np.array([26.5, 27.0, 27.4, 27.9, 28.5, 29.0])  # in °C

# Prediction year
future_year = np.array([[2050]])

# Function to train and predict
def predict_future(X, y, target_year):
    model = LinearRegression()
    model.fit(X, y)
    pred = model.predict(target_year)
    return pred[0], model

# Store results
pred_population, model_pop = predict_future(years, population, future_year)
pred_vegetation, model_veg = predict_future(years, vegetation, future_year)
pred_aqi, model_aqi = predict_future(years, aqi, future_year)
pred_temp, model_temp = predict_future(years, temperature, future_year)

# Plot function
def plot_prediction(X, y, model, ylabel, future_year, future_value, color):
    plt.figure(figsize=(6, 4))
    plt.scatter(X, y, color=color, label='Historical Data')
    plt.plot(X, model.predict(X), color='blue', label='Trend Line')
    plt.scatter(future_year, future_value, color='red', label=f'2050 Prediction: {future_value:.2f}')
    plt.xlabel("Year")
    plt.ylabel(ylabel)
    plt.title(f"{ylabel} Prediction for 2050")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plots
plot_prediction(years, population, model_pop, "Population (millions)", 2050, pred_population, 'green')
plot_prediction(years, vegetation, model_veg, "Vegetation Area (sq km)", 2050, pred_vegetation, 'green')
plot_prediction(years, aqi, model_aqi, "Air Quality Index (AQI)", 2050, pred_aqi, 'orange')
plot_prediction(years, temperature, model_temp, "Average Temperature (°C)", 2050, pred_temp, 'red')

# Print predictions
print(f"Predicted Population in 2050: {pred_population:.2f} million")
print(f"Predicted Vegetation Area in 2050: {pred_vegetation:.2f} sq km")
print(f"Predicted AQI in 2050: {pred_aqi:.2f}")
print(f"Predicted Temperature in 2050: {pred_temp:.2f} °C")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Predicted values for 2050 (from previous ML analysis)
predicted_data = {
    "Year": [2050],
    "Population (millions)": [25.88],
    "Vegetation Area (sq km)": [28.7],
    "AQI": [267.3],
    "Temperature (°C)": [29.96]
}

df_pred = pd.DataFrame(predicted_data)

# Civil Engineering Interventions (Estimated Impacts)
# % improvement based on sustainable infrastructure methods
impact_factors = {
    "Green Roofing": {"Vegetation": 0.15, "AQI": -0.05, "Temp": -0.02},
    "Urban Tree Planting": {"Vegetation": 0.25, "AQI": -0.15, "Temp": -0.03},
    "Water-Sensitive Urban Design": {"Vegetation": 0.05, "AQI": -0.10, "Temp": -0.02}
}

# Apply impacts
improved_data = df_pred.copy()
total_veg_increase = sum(v["Vegetation"] for v in impact_factors.values())
total_aqi_reduction = sum(v["AQI"] for v in impact_factors.values())
total_temp_reduction = sum(v["Temp"] for v in impact_factors.values())

improved_data["Vegetation Area (sq km)"] = df_pred["Vegetation Area (sq km)"] * (1 + total_veg_increase)
improved_data["AQI"] = df_pred["AQI"] * (1 + total_aqi_reduction)
improved_data["Temperature (°C)"] = df_pred["Temperature (°C)"] * (1 + total_temp_reduction)

# Merge data for comparison
df_comparison = pd.DataFrame({
    "Metric": ["Vegetation Area (sq km)", "AQI", "Temperature (°C)"],
    "No Action (2050)": [df_pred["Vegetation Area (sq km)"][0], df_pred["AQI"][0], df_pred["Temperature (°C)"][0]],
    "With Civil Engineering": [
        improved_data["Vegetation Area (sq km)"][0],
        improved_data["AQI"][0],
        improved_data["Temperature (°C)"][0]
    ]
})

# Plot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, metric in enumerate(df_comparison["Metric"]):
    axes[i].bar(["No Action", "With Solutions"],
                [df_comparison["No Action (2050)"][i], df_comparison["With Civil Engineering"][i]],
                color=["red", "green"])
    axes[i].set_title(metric)
    for j, val in enumerate([df_comparison["No Action (2050)"][i], df_comparison["With Civil Engineering"][i]]):
        axes[i].text(j, val + (0.5 if i != 0 else 2), f"{val:.2f}", ha="center")

plt.suptitle("Impact of Civil Engineering Solutions on Dhaka's 2050 Predictions", fontsize=16)
plt.tight_layout()
plt.show()

# Show the comparison table
print(df_comparison)
```
